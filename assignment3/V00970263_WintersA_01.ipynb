## 1.1 Dataset

```python
import numpy as np
np.random.seed(1337)
```

```python
from tensorflow.keras.datasets import fashion_mnist
```

```python
# Classes: 
# 0 - T-shirt/top
# 1 - Trouser
# 2 - Pullover
# 3 - Dress
# 4 - Coat
# 5 - Sandal
# 6 - Shirt
# 7 - Sneaker
# 8 - Bag
# 9 - Ankle boot
```

```python
# Load data
(train_X, train_y), (test_X, test_y) = fashion_mnist.load_data()

# Get random indices for each class
tops = np.random.choice(np.where(test_y == 0)[0], 500, replace=False)
trousers = np.random.choice(np.where(test_y == 1)[0], 500, replace=False)
pullovers = np.random.choice(np.where(test_y == 2)[0], 500, replace=False)
dresses = np.random.choice(np.where(test_y == 3)[0], 500, replace=False)
coats = np.random.choice(np.where(test_y == 4)[0], 500, replace=False)
sandals = np.random.choice(np.where(test_y == 5)[0], 500, replace=False)
shirts = np.random.choice(np.where(test_y == 6)[0], 500, replace=False)
sneakers = np.random.choice(np.where(test_y == 7)[0], 500, replace=False)
bags = np.random.choice(np.where(test_y == 8)[0], 500, replace=False)
ankle_boots = np.random.choice(np.where(test_y == 9)[0], 500, replace=False)

truncated_idxs = np.concatenate((tops, trousers, pullovers, dresses, coats, sandals, shirts, sneakers, bags, ankle_boots))
X_test = np.delete(test_X, truncated_idxs, axis=0)
y_test = np.delete(test_y, truncated_idxs, axis=0)

print("X_test Shape:", X_test.shape[0], "\ny_test Shape:", y_test.shape[0])

print("\nAssert Each Class Size is 500:")
for i in range(0, 10):
    print("Class {} Size: {}".format(i, len(np.where(y_test == i)[0])))
```

## 1.2 Dimensionality

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import pandas as pd
from sklearn.decomposition import PCA
```

```python
# Find components needed to explain 75% of the variance
explained_var = 0.75

# Flatten 28x28 images to 1x784
X_test = X_test.reshape(-1, 784)

pca = PCA(explained_var).fit(X_test)

# Plot the top 4 components
for i in range(0, 4):
    plt.figure(figsize=(4, 4))
    plt.title("Principal Component: {}".format(i + 1))
    plt.imshow(pca.components_[i].reshape(28, 28), cmap='bone')
    plt.show()

# Use PCA to reduce the dimensions of X_test
reduced_pca = pca.transform(X_test)

# Save output for Problem 2. 
df = pd.DataFrame(reduced_pca)
df['label'] = pd.DataFrame(y_test)
df.to_csv("fashion_mnist_pca.csv", index=False)

# Get the number of components before and after PCA
d_before = X_test.shape[1]
d = reduced_pca.shape[1]

print("\nBefore PCA Components: {} | After PCA Components: {}".format(d_before, d))
print("Therefore, {} components are needed to explain 75% of the variance\n".format(d))

# Take an approx. after converting back to 784 dimensions
pca_T = pca.inverse_transform(reduced_pca)

# Get 3 random images and compare them to reduced_pca images
idxs = np.random.choice(X_test.shape[0], 3)

for idx in idxs:
    # Original image
    plt.figure(figsize=(8, 8))
    plt.subplot(1, 2, 1)
    plt.title("Original {} Dimensional Image".format(d_before))
    image = X_test[idx].reshape(28, 28)
    plt.imshow(image, cmap='gray')
    
    # reduced_pca
    plt.subplot(1, 2, 2)
    plt.title("Projected {} Dimensional Image".format(d))
    pca_image = pca_T[idx].reshape(28, 28)
    plt.imshow(pca_image, cmap='gray')
    plt.show()
```

It looks like the top 4 components are T-shirt, trouser, pullover(maybe), and, sneaker. Furthermore, 14 components are needed to explain 75% of the variance. Therefore d = 14. 

## 1.3 Plot

```python
plt.figure(figsize=(8, 8))

colours_by_label = {0: 'blue', 1: 'green', 2: 'red', 3: 'yellow', 
                    4: 'orange', 5: 'grey', 6: 'cyan', 7: 'purple', 8: 'black', 9: 'pink'
                   }

for colour in colours_by_label:
    label_rows = y_test == colour
    plt.scatter(x=reduced_pca[label_rows, 0], y=reduced_pca[label_rows, 1], c=colours_by_label[colour], label=colour, alpha=0.7)
    

plt.xlabel("Principal Component: 1")
plt.ylabel("Principal Component: 2")
plt.legend(title="Classes", labels=['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot'])
plt.title("Top 2 PCA Principal Component 2D Scatterplot")
plt.show()
```

I would have no idea which class is which if I did not know beforehand. They are all pretty much ontop of eachother. 

## 1.4 SNE

```python
from sklearn.manifold import TSNE
```

```python
tsne = TSNE(perplexity=10, learning_rate=100, init='random', angle=0.1, verbose=1)

reduced_tsne = tsne.fit_transform(X_test)

plt.figure(figsize=(8, 8))

colours_by_label = {0: 'blue', 1: 'green', 2: 'red', 3: 'yellow', 
                    4: 'orange', 5: 'grey', 6: 'cyan', 7: 'purple', 8: 'black', 9: 'pink'
                   }

for colour in colours_by_label:
    label_rows = y_test == colour
    plt.scatter(x=reduced_tsne[label_rows, 0], y=reduced_tsne[label_rows, 1], c=colours_by_label[colour], label=colour, alpha=0.7)
    

plt.xlabel("Principal Component: 1")
plt.ylabel("Principal Component: 2")
plt.legend(title="Classes", labels=['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot'])
plt.title("Top 2 t-SNE Principal Component 2D Scatterplot")
plt.show()
```

t-SNE is more informative than the PCA-based one. However, it would likely still be quite difficult to discerne classes from one another, especially in the middle/left quadrant.
