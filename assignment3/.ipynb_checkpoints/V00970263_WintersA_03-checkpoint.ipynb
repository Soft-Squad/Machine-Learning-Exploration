```python
import numpy as np
np.random.seed(1337)
```

```python
import pandas as pd
```

```python
df = pd.read_excel('online_retail_II.xlsx')
```

```python
df.info()
```

```python
df.isnull().values.any()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().values.any()
```

```python
from apyori import apriori
```

```python
countries = np.unique(df['Country'])

# manually tested minimum support values needed for aprior to run effectively for each country (described in md below)
minsups = {'Austria': 0.2, 'Bahrain': 0.5,'Canada': 0.3,'Channel Islands': 0.19, 'Cyprus': 0.2, 'Denmark': 0.3, 
           'European Community': 0.5,'Greece': 0.4, 'Hong Kong': 0.4, 'Iceland': 0.5, 'Israel': 0.3, 'Japan':0.2, 
           'Lithuania': 0.5, 'Malta': 0.4,'Poland': 0.2, 'Singapore': 0.4,'USA': 0.4, 'United Kingdom': 0.05
          }

for country in countries:
    print("Country: {}".format(country))
    df1 = df[(df['Country'] == country) & (df['StockCode'] != 'POST') & (df['Quantity'] > 0)]
    df1['Description'] = df1['Description'].str.strip()
    
    n_invoices = len(np.unique(df1['Invoice']))
    if country == 'Unspecified':
        continue
        
    if n_invoices <= 2:
        continue
    
    if country in minsups:
        minsup = minsups[country]
    
    else:
        minsup = 0.1
    
    tsxns = list(df1.groupby('Invoice')['Description'].apply(list))
    
    asns = list(apriori(tsxns, min_support=minsup))
    
    confidence_vals = {}
    i = 0
    
    for asn in asns:
        support = asn.support
        for rule in asn.ordered_statistics:
            confidence = rule.confidence
            lift = rule.lift
            left_asc = list(rule.items_base)
            right_asc = list(rule.items_add)
            confidence_vals['record_' + str(i)] = {'confidence': confidence,
                                                   'lift': lift,
                                                   'left_asc': left_asc,
                                                   'right_asc': right_asc,
                                                   'support': support
                                                  }
            i += 1
            top5 = sorted(confidence_vals.items(), key=lambda record: record[1]['confidence'], reverse=True)[:5]
            top5 = dict(top5)
            
            for record in top5:
                confidence = top5[record]['confidence']
                support = top5[record]['support']
                lift = top5[record]['lift']
                left_asc = top5[record]['left_asc']
                right_asc = top5[record]['right_asc']
                print("Association: {} --> {} Support: {:.3f} Confidence: {:.3f} Lift: {:.3f}".format(left_asc, right_asc, support, confidence, lift))
```
