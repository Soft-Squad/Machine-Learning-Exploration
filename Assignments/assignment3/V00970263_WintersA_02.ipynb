```python
k_list = list(range(2, 16))
```

```python
def init_centroids(X, k, method='random'):
    if method == 'k-means++':
        # Initial centers
        centers = np.empty((k, X.shape[1]), dtype=np.int32)
        # Choose the first center randomly and track index of points
        center_idx = np.random.randint(X.shape[0])
        centers[0] = X[center_idx]
        
        # Calculate the distance between the center and data points
        closest_dists = get_centroid_distance(X[:, None, :], centers[0, None, np.newaxis])
        total_dist = closest_dists.sum() # Sum of distances
        
        for c in range(1, k):
            # Generate random values and then choose centers based on cumululative prob
            rand_vals = np.random.uniform(size=(2 + int(np.log(k)))) * total_dist
            center_ids = np.searchsorted(np.cumsum(closest_dists), rand_vals)
            
            # Calculate the distance between the new centers and data points
            cand_dists = get_centroid_distance(X[:, None, :],  X[center_ids])
            
            # Update closest distances and potential of each candidate
            np.minimum(closest_dists, cand_dists, out=cand_dists)
            candidates = cand_dists.sum(axis=0)
            
            # Find the best candidate
            best_cand = np.argmin(candidates)
            curr_cand = candidates[best_cand]
            closest_dists = cand_dists[best_cand] 
            best_cand = center_ids[best_cand]
            
            centers[c] = X[best_cand]
            
        return centers
            
    return X[random.sample(range(X.shape[0]), k)]
```

```python
def get_centroid_distance(X, centroid):
    # Compute the l2 difference
    return np.square(X - centroid).sum(axis=(X.ndim - 1))
```

```python
def get_closest_centroid(X, centroids):
    # Compute the distance from the data point for each centroid
    dist = get_centroid_distance(X, centroids)
    
    # Return the index of the centroid with the smallest distance to the data point
    return np.argmin(dist, axis=1)
```

```python
def get_sse(X, centroids, a_centroids):
    sse = 0
    sse = get_centroid_distance(X, centroids[a_centroids]).sum() / len(X)
    return sse
```

```python
def KMeans(X, k, method, max_iter):
    
    if method == 'k-means++':
        centroids = init_centroids(X, k, method)
        
    else: 
        # Initialize centroids by specified k, and method
        centroids = init_centroids(X, k, method)

    # List to store assigned centroids
    a_centroids = np.zeros(len(X), dtype=np.int32)
    
    for i in range(max_iter):
        # Get the closest centroid to each data point
        a_centroids = get_closest_centroid(X[:, None, :], centroids[None, :, :]) # Broadcast an 3D Arrays

        # Compute new centroids
        for c in range(k):
            # Get data points by cluster and compute mean
            cluster = X[a_centroids == c]
            if len(cluster) == 0:
                continue
            cluster = cluster.mean(axis=0)

            # Update centroids
            centroids[c] = cluster

        # Compute SSE
        sse = get_sse(X.squeeze(), centroids.squeeze(), a_centroids)
    
    return sse
```

### Random Initialization

```python
# Store returned SSE's
lloyd_sse = []
for k in k_list:
    lloyd_sse.append(KMeans(X, k, 'random', max_iter=1))

plt.figure(figsize=(8, 8))
plt.plot(k_list, lloyd_sse)
plt.xlabel('k')
plt.ylabel('Sum Squared Error')
plt.title('SSE over k = [2 - 15] with Random Initialization')
plt.show()
```

## k-means++ Initialization

```python
kmeanspp = []
for k in k_list:
    kmeanspp.append(KMeans(X, k, 'k-means++', max_iter=1))

plt.figure(figsize=(8, 8))
plt.plot(k_list, kmeanspp)
plt.xlabel('k')
plt.ylabel('Sum Squared Error')
plt.title('SSE over k = [2 - 15] with K-Means++ Initialization')
plt.show()
```

More clusters appear to generally be better, although with my K-Means++ implementation, I feel I might have something going wrong. I would choose k = 11 or 15 clusters based on the graphs. ## 2.2 Hague

```python
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
```

```python
df = pd.read_csv('fashion_mnist_pca.csv')
y = np.array(df['label'])
X = np.array(df.drop('label', axis=1))

labels = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

### Average Linkage

```python
clustering = AgglomerativeClustering(n_clusters=10, affinity='euclidean', linkage='average').fit(X)

y_pred = clustering.labels_

plt.figure(figsize=(24, 12))
colours_by_label = {0: 'blue', 1: 'green', 2: 'red', 3: 'yellow', 
                    4: 'orange', 5: 'grey', 6: 'cyan', 7: 'purple', 8: 'black', 9: 'pink'
                   }

plt.subplot(1, 2, 1)
for colour in colours_by_label:
    label_rows = y == colour
    plt.scatter(x=X[label_rows, 0], y=X[label_rows, 1], c=colours_by_label[colour], label=colour, alpha=0.7)

plt.xlabel("Component: 1")
plt.ylabel("Component: 2")
plt.legend(title="Classes", labels=labels)
plt.title("Actual Classes 2D Scatterplot")


plt.subplot(1, 2, 2)
for colour in colours_by_label:
    label_rows = y_pred == colour
    plt.scatter(x=X[label_rows, 0], y=X[label_rows, 1], c=colours_by_label[colour], label=colour, alpha=0.7)

plt.xlabel("Component: 1")
plt.ylabel("Component: 2")
plt.legend(title="Classes", labels=labels)
plt.title("Average Linkage Agglomerative Clustered Classes 2D Scatterplot")
plt.show()


n_class_items = {}

i = 0
for item in labels:
    n_class_items[item] = np.where(y_pred == i)
    i += 1
    print("Number of Points Classified as {}: {} / 5000".format(item, len(n_class_items[item][0])))

```

### Average Linkage Dendrogram

```python
plt.figure(figsize=(35, 15))
den_data = X[:, 0:2]
dendogram = sch.dendrogram(sch.linkage(den_data, method='average'))
plt.title("Average Linkage Dendrogram")
plt.show()
```

It appears that the actual classes, and the agglomerative clustered classes, that most of the agg. clustered classes do not correspond to the correct labels. Excessive overlapping implies that the problem would be very hard for any human or machine to distinguish which point belongs where. 

Furthermore, to determine the correct number of classes, a cut should be made through the longest distance that seperates two horizontal lines. This can be achieved by extending every horizontal line on the graph to the full width of the graph and then finding the largest vertical distance between neighbour horizontal lines. Hence, the cut should be made around y = 1700, implying there should be three final classes.
